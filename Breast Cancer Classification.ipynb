Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn.datasets
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

Data Collection and Processing

```python
cancer_data = sklearn.datasets.load_breast_cancer()
```

```python
cancer_data
```

```python
cancer_df = pd.DataFrame(data = cancer_data.data, columns = cancer_data.feature_names)
```

```python
cancer_df.head()
```

```python
cancer_df['label'] = cancer_data.target
cancer_df.tail()
```

```python
cancer_df.shape
```

```python
cancer_df.info()
```

```python
cancer_df.describe()
```

```python
correlation = cancer_df.corr()
correlation['label']
```

```python
plt.figure(figsize = (25,25))
sns.heatmap(correlation, cmap = 'Blues', annot = True, fmt = '.1f', cbar = True, square = True)
```

```python
cancer_df.label.value_counts()
```

1 - Benign
0 - Malignant

```python
cancer_df.groupby('label').mean()
```

```python
X = cancer_df.drop(columns = 'label', axis = 1)
y = cancer_df.label
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size = 0.2, random_state = 2)
```

```python
X.shape, X_train.shape, X_test.shape
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
train_pred = logreg.predict(X_train)
training_accuracy = metrics.accuracy_score(y_train, train_pred)
training_accuracy
```

```python
metrics.confusion_matrix(y_train, train_pred)
```

```python
test_pred = logreg.predict(X_test)
test_accuracy = metrics.accuracy_score(y_test, test_pred)
test_accuracy
```

```python
metrics.confusion_matrix(y_test, test_pred)
```
