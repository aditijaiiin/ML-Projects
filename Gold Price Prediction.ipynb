Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics
```

Data Collection and Processing

```python
gold_df = pd.read_csv('Datasets/gold_data.csv')
gold_df.head()
```

Correlation

```python
correlation = gold_df.corr()
plt.figure(figsize = (8,8))
sns.heatmap(correlation, cbar = True, square = True, fmt = '.2f', cmap = 'Blues', annot = True, annot_kws = {'size':10} )
```

```python
correlation['GLD']
```

```python
sns.distplot(gold_df['GLD'], color = 'black')
```

```python
X = gold_df.drop(columns = ['Date','GLD'], axis = 1)
y = gold_df.GLD
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 2)
X.shape, X_train.shape, X_test.shape
```

```python
model = RandomForestRegressor(n_estimators = 100)
model.fit(X_train, y_train)
```

```python
y_train_pred = model.predict(X_train)
training_score = metrics.r2_score(y_train, y_train_pred)
training_score
```

```python
y_train = list(y_train)
```

```python
plt.plot(y_train, color = 'blue', label = 'Actual Value')
plt.plot(y_train_pred, color = 'green', label = 'Predicted Value')
plt.xlabel('Number of values')
plt.ylabel('Gold Price')
plt.legend()
plt.title('Actual Price vs Predicted Price')
plt.show()
```

```python
y_test_pred = model.predict(X_test)
test_score = metrics.r2_score(y_test, y_test_pred)
test_score
```

```python
y_test = list(y_test)
```

```python
plt.plot(y_test, color = 'blue', label = 'Actual Value')
plt.plot(y_test_pred, color = 'green', label = 'Predicted Value')
plt.xlabel('Number of values')
plt.ylabel('Gold Price')
plt.legend()
plt.title('Actual Price vs Predicted Price')
plt.show()
```
